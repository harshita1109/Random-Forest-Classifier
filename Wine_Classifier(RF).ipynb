```python
for x in df.columns:
  if df[x].dtype!='object':
    sns.boxplot(df[x])
    plt.show()
```

```python
# REMOVING OUTLIERS IS AN ADDITIONAL STEP (NOT NECCESSARY)
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
  Q1=df[col].quantile(0.25)
  Q3=df[col].quantile(0.75)

  IQR=Q3-Q1

  LB=Q1-1.5*IQR
  UB=Q3+1.5*IQR

  df=df[(df[col]>LB) & (df[col]<=UB)]
  print(f"Boxplot for {col} after outlier removal:")
  sns.boxplot(df[col])
  plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['quality']=le.fit_transform(df['quality'])
df['type']=le.fit_transform(df['type'])
```

```python
X=df.drop('quality',axis=1)
X
```

```python
y=df['quality']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
result=pd.DataFrame(columns=['Actual','Predicted'])
result['Actual']=y_test
result['Predicted']=y_pred
result['Predicted'].value_counts()
```

```python
accuracy_score(y_test,y_pred)
```

**DATA IS IMBALANCED**
# So it can be corrected by using the technique of SMOT
S - Synthetic
M - Minority
O - Oversampling
T - Technique
# It helps to make the dataset balanced.

```python
trees_li=[50,75,100,125,150]

for X in trees_li:
  dummy=RandomForestClassifier(n_estimators=X)
  dummy.fit(X_train,y_train)
  pred=dummy.predict(X_test)
  print("Number of decision tress:",X,'Accuracy_score:', end=' ')
  print(accuracy_score(pred,y_test))
```

**DATA IMBALANCE:**
If our target column if there consist a class of majority like(1000-1, 60-0)

# It performs very well on the current data but it will not work well on the unseen data.

**SMOTE - Synthetic Minority Oversampling Technique**
# Used to handle the imbalance dataset.
# Technique used when a data has more of one type and very few of another.

Example:
Building a model to detect fraud
# 800 - Transactions - normal
# 50 - fraud
Therefore we can say the data is imbalanced.
It will ignore the fraud and it will only focus on the normal.

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
new_X, new_y = smote.fit_resample(X, y)
```
